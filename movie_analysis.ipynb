```python
def make_list_from_str(string):
    if isinstance(string, str):
        result_list = eval(string)
    else:
        result_list = np.nan
    return result_list
```

```python
print(type(credits.cast[0]), "　　　　castについて、str型をlist型に・・・")
credits.cast = credits.cast.map(make_list_from_str)
print(type(credits.cast[0]), "　　　　cast完了\n")

print(type(credits.crew[0]), "　　　　crewについて、str型をlist型に・・・")
credits.crew = credits.crew.map(make_list_from_str)
print(type(credits.crew[0]), "　　　　crew完了\n")

credits = credits.dropna()
credits.info()
```
